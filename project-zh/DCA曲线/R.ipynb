# 代码实现

```r
# 加载所需的包
library(rmda)
```

```r
# 读取数据
data <- read.csv("data/predict.csv")
```

```r
dc_result <- decision_curve(origin_encoder ~ predict_probability,
                            data = data,
                            fitted.risk = TRUE,
                            thresholds = seq(0, 1, by = 0.01),
                            confidence.intervals = "none")
```

```r
# 调用 summary 或 plot_decision_curve 函数来绘制图形
plot_decision_curve(dc_result)
```
